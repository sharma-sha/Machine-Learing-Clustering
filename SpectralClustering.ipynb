```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
dataset = pd.read_csv('Mall_Customers.csv')
```

```python
dataset
```

```python
#dataset: Your DataFrame.
#.iloc: The method for selecting data by integer index.
#[rows:, 3:5columns]: This slice selects all rows (:) and columns from index 3 (inclusive) up to 5 (exclusive).
#.values: Converts the selected subset to a NumPy array.
X = dataset.iloc[:, 3:5].values
```

```python
X
```

```python
from sklearn.cluster import KMeans
list1 = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', random_state = 42)
    kmeans.fit(X)
    list1.append(kmeans.inertia_)
    #import matplotlib.pyplot as plt
#range automatic count
plt.plot(range(1, 11), list1)
#tittel
plt.title('The Elbow Method')
#xlabel
plt.xlabel('Number of clusters')
#inertia
plt.ylabel('WCSS')
#reuslt
plt.show()



#listcount tarket point (5)
```

```python
list1
```

```python
from sklearn.cluster import SpectralClustering 
SpectralClustering = SpectralClustering(n_clusters=8, eigen_solver=None, n_components=None, random_state=None, n_init=10, gamma=1.0, affinity='rbf', n_neighbors=10, eigen_tol='auto', assign_labels='kmeans', degree=3, coef0=1, kernel_params=None, n_jobs=None, verbose=False)
#model ctrat and prerite
label  = SpectralClustering.fit_predict(X)
```
